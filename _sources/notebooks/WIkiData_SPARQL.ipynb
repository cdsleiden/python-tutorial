# Wikidata SPARQL

The cell below lists all the authors who have received the Nobel Prize for Literature. The wikidata id of this prize is wd:Q37922.

Try to adjust the query below to find the laureates of the following awards and prizes:

* Booker Prize (wd:Q160082)
* Pulitzer Prize (wd:Q46525)




```python
from SPARQLWrapper import SPARQLWrapper, JSON

endpoint = 'https://query.wikidata.org/sparql'
```

```python
query = """
SELECT ?author ?authorLabel ?image ?genderLabel ?nationalityLabel ?birth_date ?birth_placeLabel ?educationLabel

WHERE {
?author wdt:P1411 wd:Q37922 .
?author wdt:P31 wd:Q5 .
OPTIONAL { ?author wdt:P21 ?gender . } . 
OPTIONAL { ?author wdt:P27 ?nationality . } . 
OPTIONAL { ?author wdt:P569 ?birth_date .} . 
OPTIONAL { ?author wdt:P19 ?birth_place .} . 
OPTIONAL { ?author wdt:P69 ?education . } . 
OPTIONAL { ?author wdt:P18 ?image . } .
                                                                                                            
 SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en".}
}
"""


sparql = SPARQLWrapper(endpoint)
sparql.setQuery(query)
sparql.setReturnFormat(JSON)
data = sparql.query().convert()
```

```python
import pandas as pd
import numpy as np 

df = pd.json_normalize(data['results']['bindings'])   
```

```python
import re

ids = df['author.value']
unique = list(set(ids))

for author_id in unique:
    author_df = df[ df['author.value'] == author_id ]
    print( 'Name: ' + author_df.iloc[0]['authorLabel.value'] )
    print( author_df.iloc[0]['image.value'] )
    print( 'Gender: {}'.format( author_df.iloc[0]['genderLabel.value']) )
    print( 'Nationality: {}'.format( author_df.iloc[0]['nationalityLabel.value']) )
    dob = author_df.iloc[0]['birth_date.value']
    if re.search( r'T' , str(dob) ):
        print( 'Date of birth: ' + dob[:dob.rindex('T')] )
    print( 'Place of birth: '.format( author_df.iloc[0]['birth_placeLabel.value']) )
    
    
    edu = author_df['educationLabel.value']
    edu = edu.drop_duplicates()
    if edu.shape[0] > 0:
        for i , education in enumerate( edu.values ):
            if not( pd.isnull(education) ):
                if i == 0:
                    print('Education:')
                print(education)
        print('\n')
        
```
